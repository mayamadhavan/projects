```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('spectral.pkl')
df.head()
```

```python
name=pd.read_csv('final_df.csv')
names=name['name']
```

```python
df['name'] = names
df['like'] = 1.0
```

```python
matrix = df.pivot(columns='name', values='like').fillna(0)
matrix.head()
```

```python
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / np.sqrt(np.dot(vec1, vec1) * np.dot(vec2, vec2))
```

```python
def similar_schools(school_name, num_schools = 5, df = matrix):
    school_vector = df.loc[:, school_name]
    cosine = matrix.apply(lambda x: cosine_similarity(x, school_vector), axis=0)
    cosine = cosine.sort_values(ascending=False)
    return cosine[1:num_schools + 1]

similar_schools('Wellesley College',4)
```
